## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
import numpy as np
```

```python
mx_df= pd.read_csv('weather-raw.csv')
mx_df.head()
```

```python
mx_df.info()
```

```python
mx_df.shape
```

```python
mx_df.isna().sum()
```

```python
mx_df.drop(['d9','d12','d18','d19','d20','d21','d22'], axis= 1, inplace= True)
#Dropping this columns as all values in the column are null.
```

```python
#using melt method to convert columns into row 
mx_df = mx_df.melt(['id','year', 'month','element'],['d1', 'd2','d3','d4','d5','d6','d7','d8','d10','d11','d13','d14','d15','d16','d17','d23','d24','d25','d26','d27','d28','d29','d30','d31'], 'prev cols') 
```

```python
mx_df
```

```python
mx_df['value'].fillna(np.nan, inplace= True)
```

```python
#Dropping rows where value column has NAN
mx_df.dropna(subset=['value'])
```

```python
mx_df.to_csv('')
```
