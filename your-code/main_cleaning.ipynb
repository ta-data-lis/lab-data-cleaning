#### 3. Rename Id column to userId

#### 4. Import posts table:

```python
posts = pd.read_csv('C:/Users/migue/Documents/IRONHACK/Week 2/Labs/lab-data-cleaning/your-code/posts_table.csv')
posts.head(3)
```

#### 5. Rename Id column to postId and OwnerUserId to userId

```python
posts.rename({"Id": "postId", "OwnerUserId":"userId"}, axis = "columns", inplace = True)

posts.head(3)
```

#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python
selected_users = ["userId", "Reputation","Views","UpVotes","DownVotes"]

new_users = users[selected_users]

display(new_users.head(2))
```

```python
selected_posts = ["postId", "Score","userId","ViewCount","CommentCount"]

new_posts = posts[selected_posts]

display(new_posts.head(2))
```

#### 7. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
user_post = pd.merge(new_users, new_posts, on = "userId", how = "inner")

user_post.head(3)
```
